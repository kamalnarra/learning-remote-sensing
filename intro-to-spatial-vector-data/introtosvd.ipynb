```python
#IMPORTING PACKAGES

import os
import matplotlib.pyplot as plt
import geopandas as gpd
import earthpy as et
```

```python
#GETTING DATA READY

#setting location
directory = os.getcwd()

#getting the data
data = et.data.get_data('spatial-vector-lidar')
os.chdir(
    os.path.join(
        et.io.HOME, 
        'earth-analytics'
    )
)

#define path
plot_centroid_path = os.path.join(
    "data",
    "spatial-vector-lidar",
    "california",
    "neon-sjer-site",
    "vector_data",
    "SJER_plot_centroids.shp"
)
sjer_crop_extent_path = os.path.join(
    "data",
    "spatial-vector-lidar",
    "california",
    "neon-sjer-site",
    "vector_data",
    "SJER_crop.shp"
)

#import shapefile
sjer_plot_locations = gpd.read_file(plot_centroid_path) 
sjer_crop_extent = gpd.read_file(sjer_crop_extent_path)
```

```python
#PLOTTING DATA

#axis for points
fig, ax = plt.subplots(figsize = (10,10))

#plotting with a legend
sjer_crop_extent.plot(
    color= 'lightgrey',
    edgecolor='black',
    alpha=.5,
    ax=ax
)
sjer_plot_locations.plot(
    column = 'plot_type',
    categorical = True,
    legend = True,
    figsize = (10,6),
    markersize=45,
    cmap="Set1",
    ax=ax
)



#setting title
ax.set_title('SJER Plot Locations: Madera County, CA')


#plotting
plt.axis('equal')
plt.show()
```
